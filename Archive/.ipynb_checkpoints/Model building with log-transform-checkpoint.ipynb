```python
# Undo log transformation to get predctions in €
df_pred = pd.DataFrame((y_hat_test), columns=['Prediction'])
df_pred.head()
```

```python
df_pred['Target'] = ((y_test).reset_index(drop = True))
df_pred.head()
```

```python
df_pred['Residual'] = df_pred['Target'] - df_pred['Prediction']
```

```python
df_pred['Diff [%]'] = np.absolute(df_pred['Residual']/df_pred['Target'] * 100)
```

```python
df_pred.describe()
```

## Regression Evaluation Metrics
**Root Mean Squared Error** (RMSE)

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

```python
from sklearn.metrics import mean_squared_error

print('RMSE:',
      np.sqrt(mean_squared_error(df_pred['Target'],
                                                  df_pred['Prediction'])))
```

```python
y_true = df_pred['Target']
y_pred = df_pred['Prediction']

rmspe = (np.sqrt(np.mean(np.square((y_true - y_pred) / y_true)))) * 100

print(f'The root mean squared precent error is: \n{round((rmspe),2)}%')
```

```python
pd.options.display.max_rows = 2000
df_pred.sort_values(by = ['Diff [%]'], ascending=False).head(300)
```

```python
#lasso regression
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
lm_l = Lasso(alpha = 0.0199)
lm_l.fit(X_train, y_train)
np.mean(cross_val_score(lm_l, X_train, y_train, scoring = 'neg_root_mean_squared_error', cv = 3))
```

```python
error = []
alpha = []
for i in range(1,200):
    alpha.append(i/10000)
    lml = Lasso(alpha= (i/10000))
    error.append(np.mean(cross_val_score(lml, X_train, y_train, scoring = 'neg_root_mean_squared_error', cv = 3)))
```

```python
plt.plot(alpha, error)
```

```python
err = tuple(zip(alpha, error))
df_err = pd.DataFrame(err, columns = ['alpha', 'error'])
df_err[df_err['error'] == df_err['error'].max()]
```

## Random forest

```python
# Random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
```

```python
np.mean(cross_val_score(rf,X_train, y_train, scoring = 'neg_root_mean_squared_error', cv = 3))
```

```python
#tune models with GridsearchCV
from sklearn.model_selection import GridSearchCV

#Number of trees in random forest
n_estimators = range(295,300,5)

#Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']

#Max. number of levels in tree
max_depth = [2,5]

#Min. number of samples to split a node
max_sample_split = [2,5]

#Min. number of samples at each leaf node
min_samples_leaf = [1,3]

#Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
param_grid = { 
    'n_estimators': n_estimators,
    'max_features': max_features
}

print(param_grid)
```

```python
gs = GridSearchCV(rf, param_grid, scoring= 'neg_mean_absolute_error', cv= 3)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_score_
```
